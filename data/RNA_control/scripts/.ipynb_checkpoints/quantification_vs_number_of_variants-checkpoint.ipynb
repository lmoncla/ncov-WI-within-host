## Overview
- Plot Ct vs # of variants

Whenever possible, we include an average Ct value across N1 and N2, using the CDC QRT-PCR asssay. If we did not have both N1 and N2, we used whichever Ct was available. We did not have Ct information for 47 samples. For most of these samples, we include a relative light unit (RLU) comparison. 
 

### a note about numbering schemes 

Everything was mapped to the Wuhan reference sequence (MN908947.3).  
Nucleotide numbering scheme should match up with the numbering scheme used in Nextstrain.

```python
# load all text files in a directory into Pandas dataframes 
def number_of_variants(path):
    filelist = [os.path.splitext(file)[0] for file in os.listdir(path) if file.endswith('_intersection.csv')]
    list_of_DF_names = []
    list_of_DFs = []
    another_list = []
    for file in filelist: 
        fullname = path + '/' + file + '.csv'
        DF = pd.read_csv(fullname, header=0, sep='\t', low_memory=False)
        DF.name = file 
        list_of_DF_names.append(file)
        list_of_DFs.append(len(DF.index))
#         print(file,':',len(DF.index))
    dictionary = dict(zip(list_of_DF_names, list_of_DFs))
    return(dictionary)
#
```

```python
# run the above function 
a = number_of_variants('../data/vcfs-all')
a = {x.replace('_intersection', ''): v for x, v in a.items()} 


#
```

```python
# add a num of variants column to the dataframe with samples + cts using the dictionary we just made

DF = pd.read_csv('../data/sample_quantification.csv', sep=',')
DF['num_of_variants'] = DF['sample'].map(a)
DF = DF[['sample', 'Ct', 'RLU', 'num_of_variants']]

# DF

#
```

```python
fig, ax = plt.subplots()
ax.scatter(DF.num_of_variants, DF.RLU, color='lightblue', alpha=0.5)
# ax.axis('equal')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

ax.tick_params(axis = 'both', which = 'major', labelsize = 12)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 12)

ax.set_title('Ct vs number of variants', fontname='Arial', fontsize=12)
ax.set_ylabel('RLU', fontname='Arial', fontsize=12)
ax.set_xlabel('number of variants', fontname='Arial', fontsize=12)
for tick in ax.get_xticklabels():
    tick.set_fontname("Arial")
for tick in ax.get_yticklabels():
    tick.set_fontname("Arial")
ax.set_ylim(1000,1400)
ax.set_yticks([1000,1100,1200,1300,1400])
ax.set_xlim(-1,30)
ax.set_xticks([0,10,20,30])


# save this plot to the figures folder 
plt.savefig('../figures/variants_vs_RLU.png', bbox_inches='tight', dpi=300)

##
```
